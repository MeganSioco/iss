```python
import requests
import matplotlib.pyplot as plt
import time

def get_iss_location():
    url = "http://api.open-notify.org/iss-now.json"
    response = requests.get(url)
    data = response.json()
    timestamp = data["timestamp"]
    latitude = float(data["iss_position"]["latitude"])
    longitude = float(data["iss_position"]["longitude"])
    return timestamp, latitude, longitude

# Initialize lists to store data points
timestamps = []
latitudes = []
longitudes = []

# Generate 100 data points with 10-second intervals
for i in range(1, 101):
    timestamp, latitude, longitude = get_iss_location()
    timestamps.append(timestamp)
    latitudes.append(latitude)
    longitudes.append(longitude)
    time.sleep(10)

# Plotting the path of the ISS
plt.plot(longitudes, latitudes, marker='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of the International Space Station')
plt.grid(True)
plt.show()
```
